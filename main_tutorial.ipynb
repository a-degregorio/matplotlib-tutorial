# Learn matplotlib

Some brief examples to learn matplotlib.

```python
import matplotlib.pyplot as plt
import sklearn
import numpy as np


%matplotlib inline
```

## Plot function

Some easy examples using the function plot.

In this first example we run plot with only a list as an argument.
As we can see the abscisse of the points strats from 0 to the numer of elements in the list.

```python
plt.plot([1,2,3,4,5.5, 7.1])
plt.show()
```

With two arguments we can set the $x$ and the $y$ of the poins we want to plot.

```python
plt.plot([1,2,3,4],[5,9,2,6])
plt.show()
```

## Colors, marker, lines

We can change the apect of our plot in the same way we change the properties of a scatterplot.

At first we will explore the markers.

```python
x = np.linspace(0,10,20)
markers = '.,ov^<>1234sp*hH+xDd|_'
plt.figure(figsize= (20,10))
for i in range(len(markers)):
    plt.plot(x,2*(i+1)*np.ones(len(x)), marker = markers[i], markersize = 10, linestyle = '-', linewidth = 0)
    plt.legend(markers, loc = 2, fontsize =15)
    plt.xlim(-1,11)
plt.title('Markers example')
plt.show()
```

Now lets change linestyles:

```python
linestyles = ['-','--','-.',':']
plt.figure(figsize= (20,10))
for i in range(len(linestyles)):
    plt.plot(x,2*(i+1)*np.ones(len(x)), linestyle = linestyles[i], linewidth = 2)
    plt.legend(linestyles, loc = 2, fontsize = 20)
plt.title('Linestyles example')
plt.show()
```

## Subplots

```python
plt.figure(1, figsize=(20,10))
plt.suptitle('Ciao', fontsize = 20) # "Global" title
plt.subplot(221)
plt.plot(x,pow(x,1),'r')
plt.title('P1')
plt.subplot(222)
plt.plot(x,pow(x,2),'g')
plt.title('P2')
plt.subplot(223)
plt.plot(x,pow(x,3),'b')
plt.title('P3')
plt.subplot(224)
plt.plot(x,pow(x,4),'k')
plt.title('P4')

plt.show()
```
